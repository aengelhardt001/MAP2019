High gas temperature caused by high inflow temp or the result of it???
Ram pressure stripping can cause the overall temperature of the gas to decrease because it strips away hot gas, which then means that this hot gas cannot be recycled into the disk, potentially causing the inflow temperatures to also decrease. 

```python
plt.figure(figsize=(7,7))

plt.scatter(abs(data['Gout_T']), abs(data['G_inflow_2.5']), label='Field')
plt.scatter(abs(MW['Gout_T']), abs(MW['G_inflow_2.5']), label='satellites')

plt.loglog()
plt.xlim(2e3, 2e6)
plt.ylim(1e-6, 1e2)

plt.legend()
plt.ylabel(r'Inflow ($M_{\odot}$/yr)', size=15)
plt.xlabel('Inflow Temp (K)', size=15)
plt.title('Temp of inflow vs Rate if inflow', size=20)
```

Trend of Satellite galaxies where halos with higher inflow rates have cold gas and halos with lower inflow rates have hotter gas. Could be indicative of Ram stripping - SN ejected gas is not being reaccreted because it is being stripped by the host galaxy, thus rates of hot gas accretion are significantly lower. Also indicates a lack of strangulation because they halos are still able to accrete fresh cold gas from outisde the halo. 
Contrasts the Field halos which have a fairly constant temperature for their accreted gas over a large range of values for inflow rate.
