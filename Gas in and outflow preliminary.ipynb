```python
#Vc == ((Gravitational constant * Halo Mass)/Rvir)**(1/2)
G = 4.302*1e-3           #in pc/Msol(km/s)**2
h = 0.6776942783267969          #halo.properties['h']
Rvir = data['Rvir']*1000*h
V_c = (G* data['mass']/Rvir)**(1/2)     #km/s
```

```python
#Plot N value vs Vc

# EQ most likely all values will be less than 60km\s

nlow = 2.9*(V_c/60)**(-3.2)
nhigh =  2.9*(V_c/60)**(-1)

plt.figure(figsize=(7,7))

plt.scatter(V_c, data['G_outflow_2.5']/data['SFR'], c='b')
plt.plot(V_c, nlow, c='m')
plt.plot(V_c, nhigh, c='r')

plt.loglog()
plt.ylim(1, 5e2)
plt.xlim(2e1, 2e2)

plt.xlabel(r'$V_{c}$ (Km\s)', size=20)
plt.ylabel(r'n = $M_{out}$/SFR', size=20)
plt.title('Power law fit Muratov', size=25)
```
