```python
z0halonum = 116 # specify the halo number you're interested in (at z=0)

Gin2 = []
Gout2 = [] 
GinL = [] 
GoutL = [] 
# Gout_T = []
# Gin_T = []
Gout_T_25 = []
Gin_T_25 = []
Gout_T_67 = []
Gin_T_67 = []
distance = [] # e.g. distance from halo 1
time = []

for tstep in range(len(haloids[z0halonum])):
    snapshot = snapshots[tstep]
    
    # load in data in pynbody
    s = pynbody.load(path+snapshot)
    s.physical_units()
    h = s.halos()
        
    i = haloids[z0halonum][tstep]
    halo = h.load_copy(i)
    print('\t Major progenitor halod ID:', i)    
    
    t = float(h[i].properties['time'].in_units('Gyr'))
    print(f'Loaded snapshot {snapshot}, {13.800797497330507 - t} Gyr ago')
    
    #get halo 1 information and calculate distance from halo i to halo 1
    h1id = haloids[1][tstep] # get %equivalent halo 1 id
    h1d = np.array([h[h1id].properties['Xc'], h[h1id].properties['Yc'], h[h1id].properties['Zc']]) # position of halo 1
    h1r = h[h1id].properties['Rvir'] # virial radius of halo 1
    d = np.array([h[i].properties['Xc'], h[i].properties['Yc'], h[i].properties['Zc']]) # position of halo i
    d = np.sqrt(np.sum((d - h1d)**2))/h1r # distance between the two 
    print('\t Distance from halo 1 (technically halo %s): %.2f Rvir' % (h1id,d))
    
    #center on the halos
    pynbody.analysis.halo.center(h[i])
    pynbody.analysis.halo.center(halo)
    
    ### Gas Outflow RATE
    dL = .1*h[i].properties['Rvir']/h[i].properties['h']
    Rmag = h[i]['r']
    
    print("Halo is Centered")
    
    #select the particles in a shell 0.25*Rvir
    inner_sphere2 = pynbody.filt.Sphere(str(.2*h[i].properties['Rvir']) + ' kpc', [0,0,0])
    outer_sphere2 = pynbody.filt.Sphere(str(.3*h[i].properties['Rvir']) + ' kpc', [0,0,0])
    shell_part2 = halo[outer_sphere2 & ~inner_sphere2].gas
    
    print("Shell = shell_part2")
    
    #Perform calculations
    velocity2 = shell_part2['vel'].in_units('kpc yr**-1')
    r2 = shell_part2['pos'].in_units('kpc')
    Mg2 = shell_part2['mass'].in_units('Msol')
    r_mag2 = shell_part2['r'].in_units('kpc')
    temp2 = shell_part2['temp']

    VR2 = np.sum((velocity2*r2), axis=1)
    
    #### First, the Mas Flux within the Shell ####
    
    gin2 = []
    gout2 = []
    
    for y in range(len(VR2)):
        if VR2[y] < 0: 
            gflowin2 = np.array(((VR2[y]/r_mag2[y])*Mg2[y])/dL)
            gin2 = np.append(gin2, gflowin2)
        else: 
            gflowout2 = np.array(((VR2[y]/r_mag2[y])*Mg2[y])/dL)
            gout2 = np.append(gout2, gflowout2)
    GIN2 = np.sum(gin2)
    GOUT2 = np.sum(gout2)
    
    print("Flux in", GIN2)
    
    ##### Now the temperature of particles within the Shell ####
    
    tin2 = []
    min_2 = []
    tout2 = []
    mout_2 = []
    
    for y in range(len(VR2)):
        if VR2[y] < 0: 
            intemp2 = np.array(temp2[y]*Mg2[y])
            tin2 = np.append(tin2, intemp2)
            min2 = np.array(Mg2[y])
            min_2 = np.append(min_2, min2)
        else: 
            outemp2 = np.array(temp2[y]*Mg2[y])
            tout2 = np.append(tout2, outemp2)
            mout2 = np.array(Mg2[y])
            mout_2 = np.append(mout_2, mout2)
            
    in_2T = np.sum(tin2)/np.sum(min_2)
    out_2T = np.sum(tout2)/np.sum(mout_2)    
    GIN_T_25 = np.sum(in_2T)
    GOUT_T_25 = np.sum(out_2T)
    
    print("GIN TEMP", GIN_T_25)
    
    
    #select the particles in a shell
    inner_sphereL = pynbody.filt.Sphere(str(.6*h[i].properties['Rvir']) + ' kpc', [0,0,0])
    outer_sphereL = pynbody.filt.Sphere(str(h[i].properties['Rvir']) + ' kpc', [0,0,0])
    shell_partL = halo[outer_sphereL & ~inner_sphereL].gas
    
    #Perform calculations
    DD = .4*h[i].properties['Rvir']/h[i].properties['h']
    velocityL = shell_partL['vel'].in_units('kpc yr**-1')
    rL = shell_partL['pos'].in_units('kpc')
    MgL = shell_partL['mass'].in_units('Msol')
    r_magL = shell_partL['r'].in_units('kpc')
    tempL = shell_partL['temp']

    VRL = np.sum((velocityL*rL), axis=1)
    
    #### First, the Mas Flux within a Shell ####
    
    ginL = []
    goutL = []
    
    for y in range(len(VRL)):
        if VRL[y] < 0: 
            gflowinL = np.array(((VRL[y]/r_magL[y])*MgL[y])/DD)
            ginL = np.append(ginL, gflowinL)
        else: 
            gflowoutL = np.array(((VRL[y]/r_magL[y])*MgL[y])/DD)
            goutL = np.append(goutL, gflowoutL)   
    GIN2 = np.sum(ginL)
    GOUT2 = np.sum(goutL)
    
    ##### Now the temperature of particles within the Shell ####
    
    tin67 = []
    min_67 = []
    tout67 = []
    mout_67 = []
    
    for y in range(len(VRL)):
        if VRL[y] < 0: 
            intemp67 = np.array(tempL[y]*MgL[y])
            tin67 = np.append(tin67, intemp67)
            min67 = np.array(MgL[y])
            min_67 = np.append(min_67, min67)
        else: 
            outemp67 = np.array(tempL[y]*MgL[y])
            tout67 = np.append(tout67, outemp67)
            mout67 = np.array(MgL[y])
            mout_67 = np.append(mout_67, mout67)
            
    in_67T = np.sum(tin67)/np.sum(min_67)
    out_67T = np.sum(tout67)/np.sum(mout_67)    
    GIN_T_67 = np.sum(in_67T)
    GOUT_T_67 = np.sum(out_67T)
    VR2 = np.sum((velocity2*r2), axis=1)
    print("Shell number 2 works")
            
#     ### Gas Outflow TEMP
#     VEL = halo.gas['vel'].in_units('kpc yr**-1')
#     R = halo.gas['pos'].in_units('kpc')
#     GM = halo.gas['mass'].in_units('Msol')
#     R_mag = halo.gas['r'].in_units('kpc')
#     T = halo.gas['temp']
    
#     out_T = []
#     mm_p = []

#     VR = np.sum((VEL*R), axis=1)
    
#     for x in range(len(VR)):
#         if VR[x] > 0:
#             goutmp = np.sum(T[x]*GM[x])
#             pmout = np.sum(GM[x])
#         else: 
#             gintmp = np.sum(T[x]*GM[x])
#             pmin = np.sum(GM[x])
#     gout_T = goutmp/pmout
#     gin_T = gintmp/pmin
    

    
#     Gout_T.append(gout_T)
#     Gin_T.append(gin_T)
    Gin2.append(np.sum(gin2))
    Gout2.append(np.sum(gout2))
    GinL.append(np.sum(ginL))
    GoutL.append(np.sum(goutL))
    Gout_T_25.append(GOUT_T_25)
    Gin_T_25.append(GIN_T_25)
    Gout_T_67.append(GOUT_T_67)
    Gin_T_67.append(GIN_T_67)
    distance.append(d)
    time.append(t)
```
